```
import pandas as pd
import seaborn as sns
```

```
df_iris = sns.load_dataset('iris')
df_iris.head()
```

```
df_iris.describe()
```

```
# Diagrama de scaterr plot
sns.scatterplot(data=df_iris, x='sepal_length', y='petal_length')
```

```
# Diagrama de scaterr plot, segmentando por categoria species
sns.scatterplot(data=df_iris, x='sepal_length', y='petal_length', hue='species')
```

```
df_iris.groupby(['species']).count()
```

```
# Joint plot
# Permite ver un scater plot, y en el eje x y y muestra su compartamiento en la distribucion
sns.jointplot(data=df_iris, x='sepal_length', y='petal_length', hue='species')
```

```
# box plot por categorias 
sns.boxplot(x = 'species', y = 'sepal_length', data = df_iris)
```

```
#Barplot
sns.barplot(x = 'species', y = 'sepal_length', data = df_iris)
```
