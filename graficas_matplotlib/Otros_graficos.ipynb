```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(1, 50, 100)
print(data)
```

### Histograma

```python
plt.hist(data, bins=10, histtype='bar')
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step')
plt.show()
```

## Boxplot

```python
data = np.append(data, 200)
# plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
#  showfliers=False // ocultaria los outliers
plt.boxplot(data, notch=True, vert=False)
plt.show()
```

### Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2
colors = np.random.rand(N)

plt.scatter(x, y, s=area, c=colors, marker='o', alpha=0.5)
plt.show()
```
