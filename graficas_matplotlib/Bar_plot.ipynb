```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

```python
plt.bar(country, population, width=0.5, color=['red', 'blue', 'green', 'gold'])
# Modificando los labels en el eje x, rotandolo N grados
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation=45)
plt.show()
```

```python
# Barras horizontales
plt.barh(country, population)
plt.show()
```
