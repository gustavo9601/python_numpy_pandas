```python
# Imports
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X = ['a', 'b', 'c']
Y = [1, 2, 3]
sns.barplot(X, Y)
plt.show()
```

### Set
.set() permite configurar los estilos del grafico

```python
sns.set(style='dark', palette='muted', font='Verdana', font_scale=2)
sns.barplot(X, Y)
plt.show()
```

```python
tips = sns.load_dataset('tips')
print(tips.shape)
tips.head()
```

```python
# distribucion plot
# .displot() // escoge el mejor grafico dinamicamente para ver la distribucion
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# hue='sex' // agrupa por colores los datos
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
# hue='sex' // agrupa por colores los datos
# kind='kde  // fuerza el tipo de grafico a mostrar
# es muy usado ya que solo cambiado kind ejecuta cualquier grafico
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', alpha=.5)
plt.show()
```

```python
# histograma
# bins = numero de barras a mostrar en la agrupacion
# cumulative=True // permite acumular las barras sumando la anterior
sns.histplot(data=tips, x='tip', bins=20)
plt.show()
```

```python

# histograma
# bins = numero de barras a mostrar en la agrupacion
# cumulative=True // permite acumular las barras sumando la anterior
# stat='count' // tipo de operacion a aplicar en las Y
# multiple='dodge' // especifica como se debe graficar las barras
sns.histplot(data=tips, x='tip', bins=20, hue='sex', stat='probability', multiple='dodge')
plt.show()
```

```python
# .kdeplot () // diagrama de densidad
# shade=True // pinta el area bajo la curva
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True)
plt.show()
```

### Graficos para variables categoricas

```python
# Countplot  // cuenta
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
# variando el eje x o y, se mostrara horizontal o vertical
sns.countplot(data=tips, y='sex', hue='day')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
# catplot // escoge automaticamente cual debe ser el mejor grafico a visualizar
# kind = '' fuerza cual escoger
# es muy usado ya que solo cambiado kind ejecuta cualquier grafico
sns.catplot(x="day", hue="smoker",
            col="time",
            kind="count", data=tips)

plt.show()
```

### Graficos de relacion de variables

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center left', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
# es muy usado ya que solo cambiado kind ejecuta cualquier grafico
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', col='time')
plt.show()
```

```python
sns.pairplot(data=tips, hue='day')
plt.show()
```

### Join plot (permite unir n graficas en uno)

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True,
              marginal_kws=dict(bins=25, fill=False, multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True, kind='scatter')
plt.show()
```

### Heatmap (correlacion de variables)

```python
# devuelve la correlacion de las variables
# -1 a 1
# -1 Correlacion indirecta
# 0 No hay correlacion
# 1 Correlacion directa
tips.corr()
```

```python
sns.heatmap(data=tips.corr(), annot=True, cmap='coolwarm', linewidths=2, linecolor=)
plt.show()
```
