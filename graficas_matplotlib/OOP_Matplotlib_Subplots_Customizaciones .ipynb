```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 5, 11)
Y = X ** 2
print('X')
print(X)
print('Y')
print(Y)
```

```python
# .figure // lienczo
fig = plt.figure()

# .add_axes([position_x_lienzo, position_y_lienzo, size_x, size_y])
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes2 = fig.add_axes([0.2, 0.6, 0.4, 0.3])

# Se puede pintar N cantidad de graficos en funcion de las posiciones y tamaños de los axes
axes.plot(X, Y, color='red')

axes2.plot(Y, X, color='black')
axes2.set_facecolor('blue')
```

```python
X = np.linspace(0, 5, 11)
Y = np.sin(X)
print('X')
print(X)
print('Y')
print(Y)
```

```python
fig, axes = plt.subplots()
axes.plot(X, Y, color='blue')
```

```python
# Accediendo a diferentes access, como array
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(X, Y, 'b')
axes[1].plot(Y, X, 'r')
```

```python
# Accediendo a diferentes access, como tuplas

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(X, Y, 'b')
ax2.plot(Y, X, 'r')
```

```python
# Accediendo a los access, por matriz
fig, axes = plt.subplots(2, 4)
axes[0, 0].plot(X, np.cos(X), color='gray')
axes[0, 1].plot(X, np.sin(X), 'y')
axes[0, 2].plot(X, np.tan(X), 'r')
axes[0, 3].plot(X, np.cos(X) ** 2)

# Agrega margin entre graficos
fig.tight_layout()
```

### Contexto en el grafico
Añadiendo, etiquetas, descripcioens, titulos etc

```python
# Accediendo a diferentes access, como array
# figsize = tamaño lienzo
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 5))
# label // convenciones de variables
axes[0].plot(X, Y, 'b', label='sin(x)')
axes[1].plot(Y, X, 'r', label='cos(x)')
# .legend() permite mostrar las convensiones
# .legend(loc='lower left')  // modifica la posicion de la convension
axes[0].legend()
axes[1].legend()

# Titulos
axes[0].set_title('Relaxion X - Y')
axes[1].set_title('Relaxion Y - X')

# Labels ejes X y Y
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

### Estilos

```python
# Permite ver todas las posibles configuraciones
print(plt.style.available)
```

```python
plt.style.use('ggplot')
# le pasamos algunos de los estilos available
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
# alpha = transparencia
# linewidth = grosor linea
# linestyle= estilo linea
# markersize= tamaño marcador
# markerfacecolor= color marcador

ax.plot(X, X + 1, color='red', marker='*', markersize=20, markerfacecolor='blue', alpha=0.5, linewidth=10)
ax.plot(X, X + 2, color='#ccc', linestyle='dashed',  linewidth=7)
ax.plot(X, X + 3, color='green')
```

```python
# Probando los diferentes estilos
for i in plt.style.available:
    plt.style.use(i)
    plt.title(f"Estilo: [{i}]")
    fig, ax = plt.subplots(figsize=(5,5))
    ax.plot(X, Y)
```
