```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 5, 11)
Y = X**2
print('X')
print(X)
print('Y')
print(Y)
```

```python
# Chart basic
plt.plot(X, Y)
plt.show()
```

```python
# Cambiando el color y el marker a graficar
# https://matplotlib.org/stable/api/markers_api.html
plt.plot(X, Y, color='orange', marker='H')
plt.show()
```

```python
# Histograma
plt.hist(X)
plt.show()
```

```python
# Pie chart
plt.pie(X)
plt.show()
```

```python
# Scatter
plt.scatter(X, Y)
plt.show()
```

```python
# Boxplot o caja de bigotes para conocer la distribucion
plt.boxplot(X)
plt.show()
```
